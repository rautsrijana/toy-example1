```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
%matplotlib inline
```

```python
data = pd.read_csv("../data/testset.csv")
data.head()
```

```python
test_image = data.iloc[10]['Data']
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
image = cv2.imread('../data/testset/'+test_image)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.figure(figsize=(20,10))
```

```python
copy = image.copy()
final = cv2.bitwise_xor(image, copy)
plt.imshow(final)
```
